```python
%sql
use ecommerce_db;
```

```python
# daily revenu etable

from pyspark.sql.functions import sum, to_date, col

df_silver = spark.table("silver_ecommerce")

df_gold_daily = df_silver \
.withColumn("order_date", to_date(col("InvoiceDate"))) \
.groupBy("order_date") \
.agg(sum("total_amount").alias("daily_revenue"))

display(df_gold_daily)
```

```python
# checking the number of days of sales data we have
df_gold_daily.count()
```

```python
# saving gold daily revenue table
df_gold_daily.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_daily_sales")
```

```python
%sql
-- verifying it

SELECT * FROM gold_daily_sales ORDER BY daily_revenue DESC LIMIT 5;
```

```python
# building another business table i.e; top 10 products table
from pyspark.sql.functions import desc

df_silver = spark.table("silver_ecommerce")

df_gold_top_products = df_silver \
.groupBy("Description") \
.agg(sum("total_amount").alias("total_revenue")) \
.orderBy(desc("total_revenue")) \
.limit(10)

display(df_gold_top_products)
```

```python
# saving it
df_gold_top_products.write.format("delta") \
.mode("overwrite") \
.saveAsTable("gold_top_products")
```
